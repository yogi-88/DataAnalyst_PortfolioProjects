## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
new_df.isnull().sum().sort_values()
```

```python
## impute the missing values as per the datatype of the columns

for col in new_df.columns:
    if new_df[col].dtype == 'object':
        new_df[col] = new_df[col].fillna(new_df[col].mode().iloc[0])
    else:
        new_df[col] = new_df[col].fillna(new_df[col].median())
```

```python
## checking if we have null values
new_df.isnull().sum().sort_values()
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
new_df.shape
```

```python
new_df.duplicated().sum()
```

```python
## dropping duplicates
new_df.drop_duplicates(inplace=True)
```

```python
## Display the total number of records in the dataframe after removing the duplicates.
new_df.shape
```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
## Rename the column.
new_df.rename(columns={'availability 365':'days_booked'},inplace=True)
```

```python
new_df.head(2)
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
new_df.columns = [col.lower().replace(" ","_") for col in new_df.columns]
```

```python
new_df.columns 
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

new_df[['price','service_fee']].head()
```

```python
def remove_dollar_coma(value):
    if pd.isna(value):
        return np.NaN
    else:
        return float(value.replace("$","").replace(",",""))
```

```python
new_df['price'] = new_df['price'].apply(remove_dollar_coma)
new_df['service_fee'] = new_df['service_fee'].apply(remove_dollar_coma)
```

```python
## List the top 10 neighborhoods in the increasing order of their price with the help of a horizontal bar graph. Which is the cheapest neighborhood.
top_10_neighbourhood = new_df.groupby('neighbourhood')['price'].mean().sort_values().head(10)
plt.figure(figsize=(10,6))
top_10_neighbourhood.plot(kind='barh', color='green')
plt.title('Top 10 Neighborhoods with increasing prices')
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.show()
```

```python
#short term rentals
short_term_rentals = new_df[new_df['minimum_nights'] <= 10]['neighbourhood'].value_counts()
short_term_rentals
```

```python
plt.figure(figsize=(10,6))
short_term_rentals.plot(kind='bar', color='lightcoral')
plt.title('Neighborhoods offering short term rentals within 10 days')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.show()
```

```python
## Data visualisation to show List the prices with respect to room type using a bar graph and also state your inferences.
price_wrt_roomtype = new_df.groupby(['room_type'])['price'].mean()
price_wrt_roomtype
```

```python
plt.figure(figsize=(10,6))
price_wrt_roomtype.plot(kind='bar', color='lightcoral')
# sns.barplot(x='room_type',y='price',data=price_wrt_roomtype, palette='coolwarm')
plt.title('prices with respect to Room type')
plt.xlabel('Romm type')
plt.ylabel('Average Price')
plt.show()
```

```python
## Create a pie chart that shows distribution of booked days for each neighborhood group .Which neighborhood has the highest booking percentage.

booked_days_per_neighborhood = new_df.groupby('neighbourhood')['days_booked'].sum()

plt.figure(figsize=(8,8))
plt.pie(
    booked_days_per_neighborhood, 
    labels=booked_days_per_neighborhood.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=plt.cm.Paired.colors)
plt.title('Distribution of booked days by Neighborhood group')
plt.show()


```

```python
# Neighborhood with highest booking percentage
highest_booking_neighborhood = booked_days_per_neighborhood.idxmax()
highest_booking_neighborhood
```
